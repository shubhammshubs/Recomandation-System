Problem statement.

Build a recommender system by using cosine simillarties score.


# 1. Import Libs 

```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
```

# 2. Import Data 

```python
book_data = pd.read_csv('book.csv',encoding='Latin1')
book_data
```

# 3. Data Understanding

#### Droping (Unnamed: 0) column

```python
book_data2 = book_data.drop(['Unnamed: 0'],axis=1)
```

```python
book_data2
```

### Sort values by User IDs

```python
book_data2.sort_values(['User.ID'])
```

```python
book_data2.shape
```

#### Total number of unique users in the dataset 

```python
book_data2['User.ID'].nunique()
```

#### Total number of unique books in the dataset

```python
book_data2['Book.Title'].nunique()
```

# 4. Data Prearation

#### Using pivot table to replace long data into wide data


```python
user_pivot_table = book_data2.pivot_table(values='Book.Rating', index='User.ID',columns='Book.Title',).fillna(0)
user_pivot_table.columns = book_data2['Book.Title'].unique()
user_pivot_table
```

# 5. Model Building

## Metric = Cosine 

#### Calculating Cosine Similarity between Users on array data

```python
user_sim=1-pairwise_distances(user_pivot_table.values,metric='cosine')
user_sim
```

#### Set the index and column names to user ids 

```python
cosine_metrics_user = pd.DataFrame(user_sim)
cosine_metrics_user.index=book_data2['User.ID'].unique()
cosine_metrics_user.columns=book_data2['User.ID'].unique()
cosine_metrics_user
```

#### Most Similar Users

```python
cosine_metrics_user.idxmax(axis=1)
```
